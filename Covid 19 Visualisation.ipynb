```python
import dash
from jupyter_dash import JupyterDash
from dash import dash_table
from dash import dcc
from dash import html
import dash_bootstrap_components as dbc
import plotly.express as px
import pandas as pd
from dash.dependencies import Input, Output
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
path_data = "/content/drive/MyDrive/Mes Codes/Visualisation de Données/Projet Data Viz/Collectif/COVID-19-geographic-disbtribution-worldwide - COVID-19-geographic-disbtributi.csv"
data = pd.read_csv(path_data)
```

```python
def recupDate(val):
  return val.split("/")[2]
```

```python
def recupDate(val):
  return val.split("/")[2]
data['Annee'] = data["dateRep"].apply(recupDate)
data['Annee']
```

```python
data_cumu_cas_mois = data.groupby(['month', "countriesAndTerritories"])['cases'].sum().reset_index()
data_cumu_mort_mois = data.groupby(['month', "countriesAndTerritories"])['deaths'].sum().reset_index()
```

```python
data_top_cas = data.sort_values(['cases'], ascending=False, axis=0).head(10)
data_top_morts = data.sort_values(['deaths'], ascending=False, axis=0).head(10)
```

```python
app = JupyterDash(__name__, external_stylesheets=[dbc.themes.BOOTSTRAP])
dataCumulCasJours = data.groupby(["day", 'countriesAndTerritories'])['cases'].sum().reset_index()

app.layout = html.Div([
    html.Div([
        html.H1('Covid-19 dashboard', className="text-light", style={"padding":"25px","textAlign" : "Center"}),    
    ], className="bg-dark"),

    html.Div([   
        html.Div([
        html.Div(className="col-md-4"),

            html.Div([
                html.H4('Choisir un pays', className="", style={}), 
                dcc.Dropdown(
                id = 'pays',
                multi=True,
                options=[
                    {"label" : x, "value" : x} for x in data['countriesAndTerritories'].sort_values().unique()
                ],
                placeholder='Selectionner un pays',
                style={"width" : '500px'}
                )
            ], className="col-md-5"),

            html.Div(className="col-md-3"),
        ], className="row"),

        html.Div([
                html.H4('Choisir un date', className="", style={}), 
                dcc.Slider(
                    id = "date",
                    min = int(data['Annee'].min()),
                    max = int(data['Annee'].max()),
                    value = int(data['Annee'].min()),
                    marks = {str(x): str(x) for x in data['Annee'].unique()},
                    
                    step = None
                        )
            ]),

        html.Hr(),
    ], className=""),

    # -----------------------------------------------------
    html.Div([
        html.Div([
            html.Div([
                html.H5('Total des cas',
                        style = {"textAlign":"center"}, className=""
                        ),
                html.P(id='cas', children= f"{data['cases'].sum()} cas.",
                    style={"textAlign":"center",'color':'orange', 'fontSize':35}
                    ),
                html.P(id='pourcas', children= f"{data['cases'].sum()*100/data['cases'].sum()} %.",
                    style={"textAlign":"center",'color':'orange', 'fontSize':20}
                    ),
                # html.Div(dcc.Graph(id='bar',figure=px.bar(data,x='countriesAndTerritories',y='cases',title='Nombre de cas par pays')))
            ], className="col-lg-5 card-body bg-light rounded"),

            html.Div(className="col-lg-2"),

            html.Div([
                html.H5('Total des morts',
                        style = {"textAlign":"center"}, className=""
                        ),
                html.P(id='morts', children= f"{data['deaths'].sum()} morts.",
                    style={"textAlign":"center",'color':'red', 'fontSize':35},
                    className=""
                    ),
                html.P(id='pourmorts', children= f"{data['deaths'].sum()*100/data['deaths'].sum()} %.",
                    style={"textAlign":"center",'color':'red', 'fontSize':20},
                    className=""
                    ),
            ], className="col-lg-5 card-body bg-light rounded"),
        ], className="row"),
    ], className="container text-dark"),
    # -----------------------------------------------------
    html.Hr(),

    html.Div([
        html.Div([
            html.Div(
                dcc.Graph(
                    id='casMois',
                    figure= px.bar(data, x='month', y='cases', title='<b>nombre total de cas par mois</b>')
                    ), className="col-md-6"
            ),

            html.Div(
                dcc.Graph(
                    id='mortsMois',
                    figure=px.bar(data, x='month', y='deaths', title='<b>nombre total de morts par mois</b>')
                    ), className="col-md-6"
            ),
        ], className="row", style={'margin' : "30px"}),

    html.Div([
        html.Div(
            dcc.Graph(id='nouvCas',
                figure=px.bar(data, x="day", y='cases', title='<b>tendance des nouveaux cas par jour</b>',
                color='countriesAndTerritories', labels=dict(x='day',y='cases'))
                ),className="col-md-7"
        ),

        html.Div(
            dcc.Graph(id='cumul_cas_par_mois',
                figure=px.bar(data_cumu_cas_mois, x="month", y='cases', title='<b>cumul des cas par mois</b>', labels=dict(x='day',y='cases'))
                ),className="col-md-5"
        ),
    ], className="row", style={'margin' : "30px"}),

    html.Div([
        html.Div(
            dcc.Graph(id='cumul_cas_par_morts',
                figure=px.bar(data_cumu_mort_mois, x="month", y='deaths', title='<b>cumul des morts par mois</b>', labels=dict(x='day',y='deaths'))
                ),className="col-md-5"
        ),

        html.Div(
            dcc.Graph(id='CasJour',
                figure=px.bar(dataCumulCasJours, x="day", y='cases', title='<b>tendance des nouveaux cas par jour</b>',
                color='countriesAndTerritories', labels=dict(x='day',y='cases'))
                ),className="col-md-7"
        ),
    ], className="row", style={'margin' : "30px"}),
    ]),

    html.Hr(),

    html.Div([
      html.Div([
            html.H4("Dash data-table des pays qui ont cumuler le plus de cas", style={'color' : "white", 'textAlign' : "center"}),
            dash_table.DataTable(
            id='dash_table_cases',
            columns = [{"name": i, "id": i} for i in data_top_cas.columns],
            data = data_top_cas.to_dict('records'),
        ),                 
      ], className=''),

      html.Div([
            html.H4("Dash data-table des pays qui ont cumuler le plus de morts", style={'color' : "white", 'textAlign' : "center"}),
            dash_table.DataTable(
            id='dash_table_deaths',
            columns = [{"name": i, "id": i} for i in data_top_morts.columns],
            data = data_top_morts.to_dict('records'),
        ),
      ], className='')
    ], className='', style={'margin' : "30px"}),
    
    html.Br(),

], className="bg-dark")



@app.callback(
    [Output('cas','children'), Output('pourcas','children'), Output('casMois','figure')],
    [Input('date','value'), Input('pays','value')]
    )

def update_cases(date, pays):
    if not date or not pays:
        return f"{data['cases'].sum()} cas", f"{data['cases'].sum()*100/data['cases'].sum()} %", px.bar(data,x='month',y='cases',title='<b>Nombre de cas par pays par mois</b>')
    else:
        nb_cas = data[data['countriesAndTerritories'].isin(pays)]['cases'].sum()
        pourcas = data[data['countriesAndTerritories'].isin(pays)]['cases'].sum()*100/data['cases'].sum()
        graph = px.bar(data[data['countriesAndTerritories'].isin(pays)], x='month', y='cases',title=f'<b>Nombre de morts par mois pour {pays}</b>')
        return f"{nb_cas} cas.", f"{round(pourcas, 2)} %", graph

@app.callback(
    [Output('morts','children'), Output('pourmorts','children'), Output('mortsMois','figure')],
    [Input('date','value'), Input('pays','value')]
    )

def update_morts(date, pays):
    if not date or not pays:
        return f"{data['deaths'].sum()} morts", f"{data['deaths'].sum()*100/data['deaths'].sum()} %", px.bar(data,x='month', y='deaths', title='<b>Nombre de morts par pays par mois</b>')
    else:
        nb_morts = data[data['countriesAndTerritories'].isin(pays)]['deaths'].sum()
        pourmorts = data[data['countriesAndTerritories'].isin(pays)]['deaths'].sum()*100/data['deaths'].sum()
        graph = px.bar(data[data['countriesAndTerritories'].isin(pays)], x='month', y='deaths',title=f'<b>Nombre de morts par mois pour {pays} </b>')
        return f"{nb_morts} morts", f"{round(pourmorts, 2)} %", graph


@app.callback(
    Output('cumul_cas_par_mois','figure'),
    [Input('date','value'), Input('pays','value')]
    )

def update_cumul_cas_M(date, pays):
    if not date or not pays:
        return px.bar(data_cumu_cas_mois,x='month', y='cases', title='<b>Cumul des cas par mois</b>')
    else:
        graph = px.bar(data_cumu_cas_mois[data_cumu_cas_mois["countriesAndTerritories"].isin(pays)], x='countriesAndTerritories', y='cases',title=f'<b>Cumul des cas par mois pour {pays} </b>')
        return graph


@app.callback(
    Output('cumul_cas_par_morts','figure'),
    [Input('date','value'), Input('pays','value')]
    )

def update_cumul_morts_M(date, pays):
    if not date or not pays:
        return px.bar(data_cumu_mort_mois,x='month', y='deaths', title='<b>Cumul des morts par mois</b>')
    else:
        graph = px.bar(data_cumu_mort_mois[data_cumu_mort_mois["countriesAndTerritories"].isin(pays)], x='countriesAndTerritories', y='deaths',title=f'<b>Cumul des morts par mois pour {pays} </b>')
        return graph
```

```python
if __name__ == '__main__':
    app.run_server(port=8003, debug=True)
```
